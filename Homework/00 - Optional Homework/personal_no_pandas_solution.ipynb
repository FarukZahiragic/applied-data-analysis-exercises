```python
import csv
```

```python
reader = csv.reader(open('guns-data/interactive_data.csv'))
total_deaths = 0
suicide_deaths = 0
male_suicide_deaths = 0
homicide_deaths = 0
young_male_homicide_deaths = 0
young_black_male_homicide_deaths = 0
women_homicide_deaths = 0
for row in reader:
    if 'None selected' == row[1] and 'None selected' == row[2] and 'None selected' == row[3] and 'None selected' == row[4]:
        total_deaths = int(row[5])
    if 'Suicide' == row[1] and 'None selected' == row[2] and 'None selected' == row[3] and 'None selected' == row[4]:
        suicide_deaths = int(row[5])
    if 'Suicide' == row[1] and 'Male' == row[2] and 'None selected' == row[3] and 'None selected' == row[4]:
        male_suicide_deaths = int(row[5])
    if 'Homicide' == row[1] and 'None selected' == row[2] and 'None selected' == row[3] and 'None selected' == row[4]:
        homicide_deaths = int(row[5])
    if 'Homicide' == row[1] and 'Male' == row[2] and '15 - 34' == row[3] and 'None selected' == row[4]:
        young_male_homicide_deaths = int(row[5])
    if 'Homicide' == row[1] and 'Male' == row[2] and '15 - 34' == row[3] and 'Black' == row[4]:
        young_black_male_homicide_deaths = int(row[5])
    if 'Homicide' == row[1] and 'Female' == row[2] and 'None selected' == row[3] and 'None selected' == row[4]:
        female_homicide_deaths = int(row[5])

print(f"{suicide_deaths/total_deaths * 100:.2f} % of gun deaths in the US in the last year are sucides.")
print(f"{male_suicide_deaths/suicide_deaths * 100:.2f} % of gun suicide victims in the US in the last year are male.")
print(f"{homicide_deaths/total_deaths * 100:.2f} % of gun deaths in the US in the last year are homicides.")
print(f"{young_black_male_homicide_deaths/young_male_homicide_deaths * 100:.2f} % of male homicides victims aged 15-34 in the US in the last year are black.")
print(f"{female_homicide_deaths/homicide_deaths * 100:.2f} % of homicide victims in the US in the last year are women.")
```
